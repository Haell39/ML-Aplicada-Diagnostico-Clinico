## Modelos de Classificação

### KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# dropando colunas desnecessárias
df = df.drop(columns=['id', 'Unnamed: 32'])

# Encode a coluna 'diagnosis'
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis']) # M=1, B=0
print("\nPrimeiras 5 linhas do dataset após codificação do diagnóstico:")
print(df.head())
```

```python
# Separando as características e o alvo
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

# separando os dados em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # estratificando para manter a proporção de classes

# aqui é a parte de feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Encontrando o melhor k (geralmente 5)
best_k = 0
best_accuracy = 0

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

print(f"\nMelhor 'k' encontrado: {best_k}")
print(f"Melhor acurácia para o 'k' ideal: {best_accuracy:.4f}")
```

```python
# Treinando o KNN com o melhor k
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train_scaled, y_train)

# parte de evaluating
y_pred_best_knn = best_knn.predict(X_test_scaled)
final_accuracy = accuracy_score(y_test, y_pred_best_knn)

print(f"\nAcurácia final do modelo KNN com k={best_k}: {final_accuracy:.4f}")
```

```python
""" Exemplo de Previsões """

import numpy as np

# Usamos o primeiro exemplo do conjunto de teste
new_data_point_scaled = X_test_scaled[0].reshape(1, -1)
true_diagnosis = y_test.iloc[0] # Diagnóstico verdadeiro para comparação

# Fazer a previsão com o modelo KNN treinado
predicted_diagnosis = best_knn.predict(new_data_point_scaled)
prediction_probabilities = best_knn.predict_proba(new_data_point_scaled)
```

```python
# Exibindo  resultados da previsão

print("\n--- Exemplo de Previsão ---\n")
print(f"Características de entrada (primeiras 5 escaladas): {new_data_point_scaled[0, :5]}")
print(f"Previsão do Modelo (0='{label_encoder.inverse_transform([0])[0]}', 1='{label_encoder.inverse_transform([1])[0]}'): {predicted_diagnosis[0]}")
print(f"Probabilidade da Previsão ('{label_encoder.inverse_transform([0])[0]}', '{label_encoder.inverse_transform([1])[0]}'): {prediction_probabilities[0]}")
print(f"Diagnóstico Verdadeiro (para comparação): {true_diagnosis} ('{label_encoder.inverse_transform([true_diagnosis])[0]}')")
```

### Arvore de Decisão

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Dados preparados para a Árvore de Decisão.")
print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de X_test: {X_test.shape}")
```

```python
dtree = DecisionTreeClassifier(random_state=42)
param_grid = {
    'max_depth': [None, 5, 10, 15, 20], # None --> os nós se expandem até todas as folhas serem puras ou min_samples_leaf atingido
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(dtree, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# Obter o melhor modelo e seus parâmetros
best_dtree = grid_search.best_estimator_
best_params = grid_search.best_params_
best_accuracy_dtree = grid_search.best_score_ #  acurácia média dos treinos

print("\n--- Otimização da Árvore de Decisão Concluída ---")
print(f"Melhores Hiperparâmetros: {best_params}")
print(f"Melhor Acurácia de Validação Cruzada (treino): {best_accuracy_dtree:.4f}")
```

```python
# Avaliar o melhor modelo de Árvore de Decisão no conjunto de teste
y_pred_dtree = best_dtree.predict(X_test)
final_accuracy_dtree = accuracy_score(y_test, y_pred_dtree)

print(f"\nAcurácia Final do Modelo de Árvore de Decisão no conjunto de teste: {final_accuracy_dtree:.4f}")

print("\n--- Relatório de Classificação ---")
print(classification_report(y_test, y_pred_dtree, target_names=label_encoder.classes_))
```

```python
# Exemplo de Previsão Simples
print("\n--- Exemplo de Previsão com Árvore de Decisão ---\n")

#  X_test (dados não escalados) diretamente para a Árvore de Decisão
example_data_point = X_test.iloc[[0]] # pegando a primeira linha como DataFrame
true_diagnosis_example = y_test.iloc[0]

# Fazendo a previsão
predicted_diagnosis_example = best_dtree.predict(example_data_point)
prediction_proba_example = best_dtree.predict_proba(example_data_point)

print(f"Características de entrada (primeiras 5): {example_data_point.values[0, :5]}")
print(f"Previsão do Modelo: {label_encoder.inverse_transform(predicted_diagnosis_example)[0]}")
print(f"Probabilidade da Previsão ('{label_encoder.inverse_transform([0])[0]}', '{label_encoder.inverse_transform([1])[0]}'): {prediction_proba_example[0]}")
print(f"Diagnóstico Verdadeiro (para comparação): {label_encoder.inverse_transform([true_diagnosis_example])[0]}")
```
